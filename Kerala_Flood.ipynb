```python
import numpy as np
import pandas as pd
import json
import requests
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://keralarescue.in/data/'
data = requests.get(url)
data = json.loads(data.text)
```

```python
df = pd.DataFrame(data)
```
